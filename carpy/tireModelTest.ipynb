```python
import numpy as np
from Function import Function
from TMEasy import Tire
```

```python
radius = 0.293
Jzz = 1
cz = 190000
dfx0 = 100000
dfy0 = 80000
fxm = 3900
fym = 3650
sxm = 0.11
sym = 0.16
fxs = 3600
fys = 3600
sxs = 0.4
sys = 0.5
n2L0 = 0.18
sy0 = 0.19
syE = 0.35
lamb = 2/3
TMEasy = Tire(radius, Jzz, cz, dfx0, dfy0, fxm, fym, sxm, sym, fxs, fys, sxs, sys, sy0, syE, lamb, n2L0)
```

```python
Sx = np.linspace(-0.5, 0.5, 1001)
Fz = 3500
sy = [0, 0.11, 0.22, 0.33, 0.44, 0.55]
forces = []
for slipy in sy:
    Fx = [TMEasy.get_tire_forces(Sx[i], slipy, Fz)[0] for i in range(len(Sx))]
    forces.append(Function(Sx, Fx, 'Slip longitudinal', 'For√ßa longitudinal (N)', name='Sy={:.3F}'.format(slipy)))
forces[0].comparaNPlots2(forces[1:], title='Slip Longitudinal TMEasy')
```
